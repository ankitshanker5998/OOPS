# How is encapsulation achieved in Python?
In Python, encapsulation is achieved by bundling data and methods inside a class and using naming conventions to restrict access. Python does not enforce strict data hiding at the language level . Instead, it uses a system of underscores to signal intent and protect data from accidental modification

Public: No leading underscores. Accessible from anywhere.
Protected (_): One leading underscore. Signifies internal use. It is a convention, not an enforcement.
Private (__): Two leading underscores. Triggers name mangling to restrict direct external access

2. Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle
from it and implement the area() method in both.



```python
from abc import ABC, abstractmethod
import math

# Abstract class
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

# Circle class
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

# Rectangle class
class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

# Creating objects
circle = Circle(5)
rectangle = Rectangle(4, 6)

# Displaying areas
print("Area of Circle:", circle.area())
print("Area of Rectangle:", rectangle.area())
```

3. Implement a multi-level inheritance scenario where a class Vehicle has an attribute type. Derive a class Car
and further derive a class ElectricCar that adds a battery attribute.



```python
# Parent class
class Vehicle:
    def __init__(self, vehicle_type):
        self.type = vehicle_type

# Child class
class Car(Vehicle):
    def __init__(self, vehicle_type, brand):
        super().__init__(vehicle_type)
        self.brand = brand

# Grandchild class
class ElectricCar(Car):
    def __init__(self, vehicle_type, brand, battery):
        super().__init__(vehicle_type, brand)
        self.battery = battery

    def display(self):
        print("Vehicle Type:", self.type)
        print("Brand:", self.brand)
        print("Battery Capacity:", self.battery)

# Creating an object
ecar = ElectricCar("Car", "Tesla", "75 kWh")

# Display details
ecar.display()
```

4. Demonstrate polymorphism by creating a base class Bird with a method fly(). Create two derived classes
Sparrow and Penguin that override the fly() method.

```python
# Base class
class Bird:
    def fly(self):
        print("Birds can fly.")

# Derived class
class Sparrow(Bird):
    def fly(self):
        print("Sparrow flies high in the sky.")

# Derived class
class Penguin(Bird):
    def fly(self):
        print("Penguin cannot fly, it swims.")

# Demonstrating polymorphism
birds = [Sparrow(), Penguin()]

for bird in birds:
    bird.fly()
```

5. Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes
balance and methods to deposit, withdraw, and check balance.



```python
# Class demonstrating encapsulation
class BankAccount:
    def __init__(self, balance=0):
        self.__balance = balance   

    # Method to deposit money
    def deposit(self, amount):
        self.__balance += amount
        print("Deposited:", amount)

    # for withdraw money
    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print("Withdrawn:", amount)
        else:
            print("Insufficient balance!")

    # for check balance
    def check_balance(self):
        print("Current Balance:", self.__balance)

# Creating an object
account = BankAccount(1000)

# Performing operations
account.deposit(500)
account.withdraw(300)
account.check_balance()
```

6. Demonstrate runtime polymorphism using a method play() in a base class Instrument. Derive classes Guitar

9. Write a class Fraction with attributes numerator and denominator. Override the str method to display the

```python
# Class to represent a fraction
class Fraction:
    def __init__(self, numerator, denominator):
        self.numerator = numerator
        self.denominator = denominator

    # Override the __str__() method
    def __str__(self):
        return f"{self.numerator}/{self.denominator}"

# Creating an object
fraction = Fraction(8, 9)

# Displaying the fraction
print("Fraction:", fraction)
```

10. Demonstrate operator overloading by creating a class Vector and overriding the add method to add two
vectors.



```python
# Class demonstrating operator overloading
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # Overloading the + operator
    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    # String representation of the object
    def __str__(self):
        return f"({self.x}, {self.y})"

# Creating vector objects
v1 = Vector(2, 3)
v2 = Vector(4, 5)

# Adding vectors
v3 = v1 + v2

# Displaying the result
print("Vector 1:", v1)
print("Vector 2:", v2)
print("Sum of Vectors:", v3)
```

12. Implement a class Student with attributes name and grades. Create a method average_grade() to compute
the average of the grades.



```python
# Class definition
class Student:
    def __init__(self, name, grades):
        self.name = name
        self.grades = grades  # list of grades

    # Method to calculate average
    def average_grade(self):
        if len(self.grades) == 0:
            return 0
        return sum(self.grades) / len(self.grades)

# Creating an object
student1 = Student("Ankit", [100, 20, 70, 4502])

# Displaying average
print("Student Name:", student1.name)
print("Average Grade:", student1.average_grade())
```

14. Create a class Employee with a method calculate_salary() that computes the salary based on hours worked
and hourly rate. Create a derived class Manager that adds a bonus to the salary.

```python
# Base class
class Employee:
    def __init__(self, name, hours_worked, hourly_rate):
        self.name = name
        self.hours_worked = hours_worked
        self.hourly_rate = hourly_rate

    # Method to calculate salary
    def calculate_salary(self):
        return self.hours_worked * self.hourly_rate


# Derived class
class Manager(Employee):
    def __init__(self, name, hours_worked, hourly_rate, bonus):
        super().__init__(name, hours_worked, hourly_rate)
        self.bonus = bonus

    # Overriding salary calculation
    def calculate_salary(self):
        base_salary = super().calculate_salary()
        return base_salary + self.bonus


# Creating objects
emp = Employee("Ankit", 40, 200)
mgr = Manager("Shanker", 40, 200, 5000)

# Display salaries
print("Employee Salary:", emp.calculate_salary())
print("Manager Salary:", mgr.calculate_salary())
```

16. Create a class Animal with an abstract method sound(). Create two derived classes Cow and Sheep that
implement the sound() method.



```python
from abc import ABC, abstractmethod

# Abstract base class
class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

# Derived class 1
class Cow(Animal):
    def sound(self):
        print("Cow sounds: Moo!")

# Derived class 2
class Sheep(Animal):
    def sound(self):
        print("Sheep sounds: Baa!")

# Creating objects
cow = Cow()
sheep = Sheep()

# Calling methods
cow.sound()
sheep.sound()
```
